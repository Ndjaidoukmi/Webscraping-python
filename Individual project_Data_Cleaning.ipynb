```python
#split the distance from location
toulouse_data[['Location','Distance']] = toulouse_data.Location.apply(lambda x: pd.Series(str(x).split(", Toulouse Show on map")))
```

```python
#create news columns for metro access
toulouse_data[['Distance', 'Metro_access']] = toulouse_data.Distance.apply(lambda x: pd.Series(str(x).split("from centre")))
```

```python
toulouse_data[['Features', 'Occupancy']] = toulouse_data.Features.apply(lambda x: pd.Series(str(x).split("Sleeps:")))
```

```python
toulouse_data[['Number_reviews']] = toulouse_data.Number_reviews.apply(lambda x: pd.Series(str(x).replace("reviews","")))
toulouse_data[['Number_reviews']] = toulouse_data.Number_reviews.apply(lambda x: pd.Series(str(x).replace(",","")))
```

```python
toulouse_data[['Features']] = toulouse_data.Features.apply(lambda x: pd.Series(str(x).replace("\n","")))
toulouse_data[['Features']] = toulouse_data.Features.apply(lambda x: pd.Series(str(x).replace(" – ","")))
```

```python
toulouse_data[['Distance','Measurement']] = toulouse_data.Distance.apply(lambda x: pd.Series(str(x).split()))
```

```python
toulouse_data[['Longitude','Latitude']] = toulouse_data.Lat_Long.apply(lambda x: pd.Series(str(x).split(",")))
```

```python
toulouse_data.head()
```

```python
toulouse_data.Latitude = toulouse_data.Latitude.astype(float)
```

```python
toulouse_data.Longitude = toulouse_data.Longitude.astype(float)
```

```python
toulouse_data.Distance = toulouse_data.Distance.astype(float)
```

```python
#cleaning price, remove euros sign
def remove_euros_signs(price):
    price_val = re.findall('[0-9]',price)
    return "".join(price_val)

toulouse_data.Price = toulouse_data.Price.apply(remove_euros_signs)
```

```python
toulouse_data.Price = toulouse_data.Price.astype(int)
```

```python
#convert the distance into km using a for loop
for i in range(len(toulouse_data)):
    if toulouse_data.loc[i, "Measurement"] == "m":
        toulouse_data.loc[i, "Distance_km"] = toulouse_data.loc[i, "Distance"]/1000
    else:
        toulouse_data.loc[i, "Distance_km"] = toulouse_data.loc[i, "Distance"]
```

```python
##Add difference categories for distance
for i in range(len(toulouse_data)):
    if toulouse_data.loc[i, "Distance_km"] <= 2:
        toulouse_data.loc[i, "Distance_classification"] = '2_km'
        
    elif toulouse_data.loc[i, "Distance_km"] <= 4:
        toulouse_data.loc[i, "Distance_classification"] = '4_km'
        
    elif toulouse_data.loc[i, "Distance_km"] <= 6:
        toulouse_data.loc[i, "Distance_classification"] = '6_km'
        
    elif toulouse_data.loc[i, "Distance_km"] <= 8:
        toulouse_data.loc[i, "Distance_classification"] = '8_km'
    else:
        toulouse_data.loc[i, "Distance_classification"] = 'Far'
```

```python
#remove colums unnecessary colums from data frame
del toulouse_data['Distance']
del toulouse_data['Measurement']
del toulouse_data['Lat_Long']
```
